```python
import requests
```

```python
page=1
url = "http://oscar-demo.eurodyn.com/surface/rest/api/search/station?programAffiliation=GOS"
 
while True:
    new_url = url + "&page={}&items=1000".format(page)
    print("getting: ",new_url)
    r = requests.get(new_url)
    
    if r.status_code != 200:
        raise Exception("wrong code")
        
    j = r.json()
    
    current_page = int(j["pageNumber"])
    page_count = int(j["pageCount"])
    total = int(j["totalCount"])
    
    print(current_page,"vs",page_count,"total:",total)
    
    if current_page == page_count:
        break
    page+=1
```

```python
%%timeit -n 1 -r 1
page=1
items = 50
url = "http://oscar-demo.eurodyn.com/surface/rest/api/search/station?programAffiliation=GOS&items={items}&page={page}"
 
while page<= 100:
    new_url = url.format(items=items,page=page)
    print("getting: ",new_url)
    r = requests.get(new_url)
    
    if r.status_code != 200:
        raise Exception("wrong code")
        
    j = r.json()
    
    current_page = int(j["pageNumber"])
    page_count = int(j["pageCount"])
    
    print(current_page,"vs",page_count)
    
    if current_page == page_count:
        break
    page+=1
    
```
